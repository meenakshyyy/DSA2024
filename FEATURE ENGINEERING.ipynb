# LINE GRAPH

```python
import matplotlib.pyplot as plt
import numpy as np

x=[0,1,2,3,4,5]
y=[0,3,2,22,22,10]
plt.plot(x,y)
plt.xlabel('days')
plt.ylabel("counts")
plt.ylim(-5,5)

plt.show()
```

```python
x=[0,1,2,3,4,5]
y1=[0,3,2,22,22,10]
y2=[1,7,5,4,3,8]
plt.plot(x,y1,c='g',label="location1")
plt.plot(x,y2,c='r',label='location2')
plt.xlabel('days')
plt.ylabel("counts")

plt.legend(loc="lower right")
plt.show()
```

# HISTOGRAMS

```python
import pandas as pd
data=pd.read_csv("auto-mpg.csv")
```

```python
plt.hist(data["mpg"],color='cyan',rwidth=0.8,density=True)
plt.title("Histogram of MPG")
plt.xlabel("mpg")
plt.ylabel("density")
plt.xticks([5,10,15,20,25,30,35,40,45,50])
plt.yticks()
plt.show()
```

```python
plt.hist(data["mpg"],color='cyan',rwidth=0.8,density=True,cumulative=True)
plt.title("Histogram of MPG")
plt.xlabel("mpg")
plt.ylabel("density")
plt.xticks([5,10,15,20,25,30,35,40,45,50])
plt.yticks()
plt.show()
```

* colinearity

```python
from scipy import stats
import matplotlib.pyplot as plt
data = np.random.exponential(scale=1,size=1000)
transformed_data,lamda_value = stats.boxcox(data)
plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
plt.hist(data,bins=30)
plt.title('Original data;')

plt.subplot(1,2,2)
plt.hist(transformed_data,bins=30)
plt.title("Box-cox transformed data")
plt.show()
```

#### Feature binning: A feature engineering technique that groups different numerical subranges into bins or buckets.
#### Feature slicing: Breaking down a product's features into separate components, or "slices," for independent development.

### Mathematical transforms:
#### Log transform: (Will remove tailedness) Log transformation is a data transformation technique that replaces each variable with its logarithm.In data with highly skewed distributions, log transformation can make patterns more visible. In data with highly skewed distributions, log transformation can make patterns more visible .

#### Box cox transform:The Box-Cox transformation is a statistical tool that transforms non-normal data into a normal distribution. This transformation can improve the accuracy of predictions made using linear regression.
